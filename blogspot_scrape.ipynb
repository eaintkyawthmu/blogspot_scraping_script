```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-infobars")
chrome_options.add_argument("--disable-dev-shm-usage")

# Set the driver
webdriver_service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

# Open the website
driver.get("https://horkhamit.blogspot.com/") # Change this to the website you want to scrape

# Create a DataFrame with the scraped text
food_df = pd.DataFrame
```

```python
# Get the HTML content of the page
html_content = driver.page_source

soup = BeautifulSoup(html_content, 'html.parser')

# Extract h3 title
title = soup.find('h3', class_='post-title entry-title').get_text(strip=True)
if title:
        # Extract h3 title
        print(f"Title: {title}")
else:
    print("Title not found")

# Extract img src under the separator class
separator = soup.find('div', class_='separator')
if separator and separator.find('img'):
    img_src = separator.find('img')['src']
    print(f"Image Source: {img_src}")
else:
    print("Image source not found")

# Extract post-body entry content and strip <br>
post_body = soup.find('div', class_='post-body entry-content')
if post_body:
    for br in post_body.find_all("br"):
        br.replace_with("\n")  # Replace <br> with newline
        # Scrape list texts
    lists = post_body.find_all('ul')
    for ul in lists:
        list_items = ul.find_all('li')
        for li in list_items:
            list_text = li.get_text(strip=True)
            print(f" - {list_text}")

    # Scrape table texts
    tables = post_body.find_all('table')
    for table in tables:
        rows = table.find_all('tr')
        for row in rows:
            cells = row.find_all('td')
            for cell in cells:
                cell_text = cell.get_text(strip=True)
                print(f"Table Cell: {cell_text}")

    post_body_text = post_body.get_text(strip=True).replace("Share:", "")
    print(f"Post Body: {post_body_text}")
else:
    print("Post body not found")
```

```python
# Create a DataFrame with the extracted data
data = {'Title': [title], 'HowTo': [post_body_text], 'Image Source': [img_src]}
food_df = pd.DataFrame(data)
```
